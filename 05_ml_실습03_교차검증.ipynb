```python
# 전처리 - 사이킥-런
from sklearn.preprocessing import LabelEncoder, OneHotEncoder

df1 = df.copy()
# 라벨 인코딩
lb_encoder = LabelEncoder()
lb_encoder.fit(df1['species'])
labels = lb_encoder.transform(df1['species'])

display(lb_encoder.classes_)
df1['label'] = labels
display(df1.head(5))

```

```python
# 원샷 인코딩
df2 = df.copy()
# 2차원 변환
labels = labels.reshape(-1, 1)

oh_encoder = OneHotEncoder()
oh_encoder.fit(labels)
oh_labels = oh_encoder.transform(labels)

# 데이터 프레임 만들기
df3 = pd.DataFrame(oh_labels.toarray(), columns=lb_encoder.classes_, dtype=int)
display(df3)

df4 = pd.concat([df2, df3], axis=1)
display(df4.head(5))
```
